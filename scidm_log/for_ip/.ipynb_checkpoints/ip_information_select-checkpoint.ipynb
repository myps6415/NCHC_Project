```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('/Users/fish/Downloads/ipDB.db')
c = conn.cursor()

ip_df = pd.read_sql('select * from important_information', conn)
ip_df
```

```python
amazon_df = ip_df[ip_df['ip_detail_column'].str.contains("amazon")]
google_df = ip_df[ip_df['ip_detail_column'].str.contains("google")]
```

```python
import matplotlib.pyplot as plt

labels = ['Amazon','Google']
size = [sum(amazon_df['count']), sum(google_df['count'])]
plt.pie(size , labels = labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
labels = ['Amazon','Google', 'Others']
size = [sum(amazon_df['count']), sum(google_df['count']), sum(ip_df['count'])-sum(amazon_df['count'])-sum(google_df['count'])]
plt.pie(size , labels = labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
