```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('/Users/fish/Downloads/ipDB.db')
c = conn.cursor()

ip_df = pd.read_sql('select * from dataset where bot is Null', conn)
ip_df
```

```python
ip_count = pd.DataFrame({'count' : ip_df.groupby(['ip']).size()}).reset_index().sort_values('count', ascending=False)
ip_count.to_sql('ip_count', conn, if_exists='replace', index = False)
```

```python
ip_count_10000 = ip_count.head(10000)
ip_count_10000
```

```python
ip_detail = pd.read_sql("select * from ip_detail where ip_detail_column <> ''", conn)
ip_detail
```

```python
df3 = ip_count_10000.join(ip_detail.set_index('ip'), on='ip')
df3
```

```python
important_information = df3.dropna()
important_information.to_sql('important_information', conn, if_exists='replace', index = False)
```
