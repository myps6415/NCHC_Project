```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3

conn = sqlite3.connect('ipDB.db')
c = conn.cursor()

url = 'https://rms.twnic.net.tw/twnic/User/Member/Search/main7.jsp?Order=ORG.ID'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
df = []

for i in soup.find('table', align= 'center').find_all('tr'):
    toDF = [s for s in i.stripped_strings]
    df.append(toDF)

df = pd.DataFrame(df)
df.columns = df.iloc[0]
df = df.drop(df.index[[0]])
df = df.drop(df.columns[-1], axis=1)
df
```

```python
df.to_sql('ipTable', conn, if_exists='replace', index = False)
```
