```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

df = pd.read_csv('/Users/fish/Downloads/dataset.csv', header=None, names=['ip', 'dataset','download', 'date', 'size', 'request', 'user_agent'])

url = 'https://raw.githubusercontent.com/monperrus/crawler-user-agents/master/crawler-user-agents.json'
df
```

```python
res = requests.get(url)

pattern = []
instances = []
for i in res.json():
    for j in i['instances']:
        pattern.append(i['pattern'])
        instances.append(j)

bot_data = pd.DataFrame(zip(pattern, instances), columns= ['bot', 'user_agent'])
bot_data
```

```python
df3 = df.join(bot_data.set_index('user_agent'), on='user_agent')
df3
```

```python
count_df = pd.DataFrame(df3['bot'].value_counts())
count_df
```

```python
import sqlite3

conn = sqlite3.connect('/Users/fish/Downloads/ipDB.db')
c = conn.cursor()
```

```python
ip_data = pd.read_sql('select ip from dataset where bot is Null', conn)
ip_data = pd.DataFrame(list(set(ip_data['ip'])), columns=['ip'])
ip_data
```

```python
ip_data['ip'][0]
```

```python
import socket

ip_detail = []
for ip in ip_data['ip']:
    try:
        aname, aip, ip = socket.gethostbyaddr(ip)
        ip_detail.append(aname)
    except:
        ip_detail.append('')

ip_data['ip_detail'] = ip_detail
ip_data
```

```python
ip_data.to_sql('ip_detail', conn, if_exists='replace', index = False)
```
