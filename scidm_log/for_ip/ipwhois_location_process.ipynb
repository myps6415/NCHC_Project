```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('/Users/fish/Downloads/ipDB.db')
c = conn.cursor()

ipwhois_table = pd.read_sql('select * from ipwhois', conn)
ipwhois_table.head()
```

```python
location_list = []
ip_detail_list = []

for detail in ipwhois_table['ip_detail_column']:
    try:
        location_list.append(detail[-2:])
        ip_detail_list.append(detail[:-4])
    except:
        location_list.append(None)
        ip_detail_list.append(None)

ipwhois_table['location'] = location_list
ipwhois_table['ip_detail_column'] = ip_detail_list
ipwhois_table.head()
```

```python
ipwhois_table.to_sql('ipwhois2', conn, if_exists='replace', index = False)
```
