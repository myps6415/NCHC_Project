```python
import pandas as pd
import datetime
import time
from ckanapi import RemoteCKAN

scidm = RemoteCKAN('https://scidm.nchc.org.tw', apikey='APIKEY')
```

```python
dataset = pd.read_csv('dataset.csv', header=None, names=['ip', 'dataset','download', 'date', 'size', 'request'])
dataset_org = pd.read_csv('dataset-organization.csv', header=None, names = ['org', 'ip', 'dataset', 'download', 'date', 'size', 'requests'])
dataset_group = pd.read_csv('dataset-group.csv', header=None, names=['group', 'ip', 'dataset','download', 'date', 'size', 'request'])
```

```python
dataset['date'] = pd.to_datetime(dataset.date, format = '%d/%b/%Y:%H:%M:%S +0800')
dataset.head()
```

```python
dataset0501 = dataset[(dataset['date'] < '2019-05-01') & (dataset['dataset'] != 'False')]
dataset0501.sort_values(by=['date'], ascending=False)
```

```python
dataset_count = pd.DataFrame({'count' : dataset0501.groupby(['dataset']).size()}).reset_index().sort_values('count', ascending=False)
dataset_count
```

```python
dataset_name_list = []
private_list = []

for data in list(dataset_count['dataset']):
    dataset_name_list.append(scidm.call_action('package_show', {'id': data})['title'])
    private_list.append(scidm.call_action('package_show', {'id': data})['private'])
    
dataset_count['dataset'] = dataset_name_list
dataset_count['private'] = private_list

dataset_count
```

```python
dataset_count_down = pd.DataFrame({'count' : dataset_0501_download.groupby(['dataset']).size()}).reset_index().sort_values('count', ascending=False)
dataset_count_down.head()
```
