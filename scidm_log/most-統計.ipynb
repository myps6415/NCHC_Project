```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# from ipwhois import IPWhois
```

```python
organization_data = pd.read_csv('log/dataset-organization.csv', header=None, names = ['org', 'ip', 'dataset', 'download', 'date', 'size', 'requests','agent'])
organization_data['date'] = pd.to_datetime(organization_data['date'], format= '%d/%b/%Y:%H:%M:%S +0800', utc=True)
#organization_data.sort_values('date', inplace=True)
#orgdata_range = organization_data[(organization_data['date'] > '2019-01-01') & (organization_data['date'] < '2019-08-02')]
orgdata_range1 = organization_data[(organization_data['date'] < '2019-01-01')]
orgdata_range4 = organization_data[(organization_data['date'] < '2019-04-01')]
orgdata_range7 = organization_data[(organization_data['date'] < '2019-07-01')]
```

```python
all_org_count_range1 = pd.DataFrame({'count1' : orgdata_range1.groupby(['org']).size()}).reset_index()
all_org_count_range4 = pd.DataFrame({'count4' : orgdata_range4.groupby(['org']).size()}).reset_index()
all_org_count_range7 = pd.DataFrame({'count7' : orgdata_range7.groupby(['org']).size()}).reset_index()
#all_org_count_range1.org
result = pd.concat([all_org_count_range1, all_org_count_range4, all_org_count_range7], axis=1, sort=False)
#result.to_csv('/home/thomas/ww/scidm_log/Result_all_browse.csv')
result
```

```python
#orgdata_range.loc[orgdata_range['org'] == 'aidm']
orgdata_range1_u = orgdata_range1.groupby(['org'])['dataset'].nunique()
orgdata_range4_u = orgdata_range4.groupby(['org'])['dataset'].nunique()
orgdata_range7_u = orgdata_range7.groupby(['org'])['dataset'].nunique()
result = pd.concat([orgdata_range1_u, orgdata_range4_u, orgdata_range7_u], axis=1, sort=False)
result
```

```python
orgdata_range_download1 = orgdata_range1[orgdata_range1['download'] == True]
orgdata_range_download4 = orgdata_range4[orgdata_range4['download'] == True]
orgdata_range_download7 = orgdata_range7[orgdata_range7['download'] == True]
all_org_count_range_download1 = pd.DataFrame({'count' : orgdata_range_download1.groupby(['org']).size()}).reset_index().sort_values('count', ascending=False)
all_org_count_range_download4 = pd.DataFrame({'count' : orgdata_range_download4.groupby(['org']).size()}).reset_index().sort_values('count', ascending=False)
all_org_count_range_download7 = pd.DataFrame({'count' : orgdata_range_download7.groupby(['org']).size()}).reset_index().sort_values('count', ascending=False)
result = pd.concat([all_org_count_range_download1, all_org_count_range_download4, all_org_count_range_download7], axis=1, sort=False)
result
```
