```python
import pandas as pd
from ckanapi import RemoteCKAN
import datetime
import time

scidm = RemoteCKAN('https://scidm.nchc.org.tw', apikey='API_KEY')

dataset = pd.read_csv('/Users/fish/Downloads/dataset.csv', header=None, names=['ip', 'dataset','download', 'date', 'size', 'request', 'user-agent'])
dataset_org = pd.read_csv('/Users/fish/Downloads/dataset-organization.csv', header=None, names = ['org', 'ip', 'dataset', 'download', 'date', 'size', 'requests', 'user-agent'])
dataset_group = pd.read_csv('/Users/fish/Downloads/dataset-group.csv', header=None, names=['group', 'ip', 'dataset','download', 'date', 'size', 'request', 'user-agent'])
```

```python
organization_data = dataset_org[dataset_org['org'] != 'False']
organization_data.head()
```

## 組織瀏覽次數

```python
org_count = pd.DataFrame({'count' : organization_data.groupby(['org']).size()}).reset_index().sort_values('count', ascending=False)
org_count
```

```python
req_list = []

for i in org_count['org']:
    try:
        req_list.append(scidm.call_action('organization_show', {'id': i})['title'])
    except:
        req_list.append(i)
        pass

org_count['org'] = req_list
org_count
```

```python
org_count.to_csv('/Users/fish/Desktop/org_count.csv', index= False)
```
