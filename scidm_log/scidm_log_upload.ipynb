```python
log_data = []
for i in data['log']:
    dt = str(i).split(' ')
    ip = re.findall(r'[0-9]+(?:\.[0-9]+){3}', dt[0])[0]
    time = (dt[3] + dt[4]).strip('[]')
    post_get = dt[5].strip('"')
    from_user = (dt[6] + ' ' + dt[7]).strip('"')
    status_code = dt[8]
    what_code = dt[9]
    url = dt[10].strip('"')
    log_data.append((ip, time, post_get, from_user, status_code, what_code, url))
    
print(log_data)
```
