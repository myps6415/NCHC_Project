```python
import pandas as pd
import datetime
import time
import random
from ckanapi import RemoteCKAN

scidm = RemoteCKAN('https://scidm.nchc.org.tw', apikey='API_KEY')
```

```python
dataset = pd.read_csv('dataset.csv', header=None, names=['ip', 'dataset','download', 'date', 'size', 'request'])
dataset_org = pd.read_csv('dataset-organization.csv', header=None, names = ['org', 'ip', 'dataset', 'download', 'date', 'size', 'requests'])
dataset_group = pd.read_csv('dataset-group.csv', header=None, names=['group', 'ip', 'dataset','download', 'date', 'size', 'request'])
```

```python
dataset['date'] = pd.to_datetime(dataset.date, format = '%d/%b/%Y:%H:%M:%S +0800')
```

```python
# 自然司
select_dataset = dataset[(dataset['dataset'] == '8541d765-1e96-4d8a-8b1d-b540f2441476') & (dataset['date'] > '2019-03-31')& (dataset['date'] < '2019-05-01')]
select_dataset.sort_values('date', ascending=True)

# ip_count = pd.DataFrame({'count' : select_dataset.groupby(['ip']).size()}).reset_index().sort_values('count', ascending=False)
# ip_count
# ip_count.to_csv('自然司瀏覽ip.csv', index= False)
```

```python
nature_df = pd.DataFrame({'count' : select_dataset.groupby(['request']).size()}).reset_index().sort_values('count', ascending=False)
nature_df
```

```python
req_list = []
for i in nature_df['request']:
    try:
        req_list.append(scidm.call_action('resource_show', {'id': i})['name'])
    except:
        req_list.append('just view outside')
        pass

nature_df['request'] = req_list
nature_df
```

```python
dataset_count = pd.DataFrame({'count': select_dataset.groupby(['dataset']).size()}).reset_index().sort_values('count', ascending=False)

# 將原本 dataset id 單獨拉出來成一個 list
dataset_list = list(dataset_count['dataset'])

#建立空 list 存 for 迴圈向 scidm 取回的資料集名稱
dataset_name = []
for dts in dataset_list:
    name = scidm.call_action('package_show', {'id': dts})['title']
    dataset_name.append(name)

# 將名稱回填到 DataFrame 的 dataset 欄位中
dataset_count['dataset'] = dataset_name
dataset_count
```
