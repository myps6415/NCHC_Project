```python
dep_datasets_count = {}
for k, v in definecate.items():
    pkg_data = ckan.action.organization_show(id=k)
    print("{2} : {0}, count {1}".format(k, pkg_data['package_count'], v))
    if v not in dep_datasets_count:
        dep_datasets_count[v] = 0
    dep_datasets_count[v] += pkg_data['package_count']
    

display(dep_datasets_count)
```
